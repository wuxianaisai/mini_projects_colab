Under-Sampling

Создание набора данных, содержащего аналогичное распределение обычной транзакции и мошеннической транзакции

```python
legit_sample = legit.sample(n=146)
```

Объединение двух фреймов данных

```python
new_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df["Class"].value_counts()
```

```python
new_df.groupby("Class").mean()
```

**3. Разделение данных на обучающую и тестовую выборки**

```python
X = new_df.drop(columns="Class", axis=1)
y = new_df["Class"]
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

**4. Обучение модели: Logistic Regression**

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_train_predict = model.predict(X_train)
training_data_acc = accuracy_score(X_train_predict, y_train)
```

```python
print(training_data_acc)
```

```python
X_test_predict = model.predict(X_test)
test_data_acc = accuracy_score(X_test_predict, y_test)
```

```python
print(test_data_acc)
```
